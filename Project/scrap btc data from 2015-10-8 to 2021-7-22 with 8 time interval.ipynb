```python
btc_time_close_volume = _1min[['Date','Close','Volume']] #we only use date close volume, therefore extract these 3 columns only
btc_time_close_volume.reset_index(drop=True, inplace=True)  #reset index
btc_time_close_volume #check
```

```python
btc_time_close_volume.to_csv('btc_time_close_volume.csv') #save #just regard this as a checkpoint
```

```python
# time in min, used for slicing
_1d=24*60
_12h=12*60
_6h=6*60
_3h=3*60
_1h=60
_30m=30
_10m=10
_1m=1 #not for analyzing, but for the nan row, will see that below
```

```python
asc_p=list(btc_time_close_volume['Close'])[::-1] #reverse the time order, more readable, asc = ascending
asc_p_arr = np.array(asc_p) #change to array first so is faster
```

```python
def cal_returns(time, name): #_p is price #function used for calculating returns
    shift_p = asc_p_arr[time:] #shifted price list
    now_p = asc_p_arr[:len(asc_p)-time] #now price list
    return_list = np.multiply(np.divide(np.subtract(shift_p, now_p), now_p),100) #new-old/old*100% formula for returns
    return pd.DataFrame({'return% '+name : return_list}) #return df
```

```python
#put together

# ~5sec
returns_df = pd.concat( #concat all returns column into one df first
    [
    cal_returns(_1m,'1m'),
    cal_returns(_10m,'10m'),
    cal_returns(_30m,'30m'),
    cal_returns(_1h,'1h'),
    cal_returns(_3h,'3h'),
    cal_returns(_6h,'6h'),
    cal_returns(_12h,'12h'),
    cal_returns(_1d,'1d'),
    ],
    axis=1
)
```

##### now start vol of diff time intervals

```python
asc_v_np=np.array(btc_time_close_volume['Volume'])[::-1] #reverse the time order, more readable, asc = ascending
#p is price, v is vol
```

```python
#this cell just for demo, nothing related to btc
#we have taklked about this in presentation already

print(np.arange(10),end='\n\n')
testingtestingtesting = np.lib.stride_tricks.sliding_window_view(np.arange(10),5)
print(testingtestingtesting)
np.sum(testingtestingtesting ,axis=1)
# same for np.std, i dont show again here
```

```python
def cal_volume(time, name): #_v is price #function used for calculating volume within time intervals
    get_all_shifted_2Darray= np.lib.stride_tricks.sliding_window_view(asc_v_np, time) #hard to explain, just see the above cell and output #btw, i show this in presentation already
    sumed = np.sum(get_all_shifted_2Darray ,axis=1) #sum elements in each inner array
    return pd.DataFrame({'volume '+name : sumed}) #return df
```

```python
#put together

# ~5sec
volume_df = pd.concat( #concat everything of the volume part
    [
    cal_volume(_1m,'1m'),
    cal_volume(_10m,'10m'),
    cal_volume(_30m,'30m'),
    cal_volume(_1h,'1h'),
    cal_volume(_3h,'3h'),
    cal_volume(_6h,'6h'),
    cal_volume(_12h,'12h'),
    cal_volume(_1d,'1d'),
    ],
    axis=1 #1 means concat along columns
)
```

no need to add one row, because volume in 1min is not return in 1min, no need to get future data, therefore not all NaN in first row, and no auto remove row will happen

then just do the same thing same as above, except adding rows

```python
volume_df_back = volume_df[::-1] #reverse back
volume_df_back.reset_index(drop=True, inplace=True) #reset index
volume_df_back.head() #check head
```

##### now start volatility(sd of return) of diff time intervals

```python
# # actually everything is the same, just different in getting
asc_r_1m_arr=np.array(returns_df_back['return% 1m'])[::-1] #reverse the time order, more readable, asc = ascending

def cal_volat(time, name): #function used for calculating volume within time intervals #time is what interval you want, name is the column name
    get_all_shifted_2Darray= np.lib.stride_tricks.sliding_window_view(asc_r_1m_arr, time) #same as the way we do for volume
    stded = np.std(get_all_shifted_2Darray ,axis=1) #we use std, because we want to get the sd within a time interval
    return pd.DataFrame({'volat '+name : stded}) #return df
```
